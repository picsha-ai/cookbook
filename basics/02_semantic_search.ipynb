# Picsha AI: Semantic Search
This notebook demonstrates how a researcher can query the Picsha AI platform using natural language to retrieve highly relevant images.

```python
import os
import requests
import json
from IPython.display import display, HTML

API_URL = "https://api.picsha.ai/v1"
API_KEY = os.getenv("PICSHA_API_KEY", "YOUR_API_KEY_HERE")
HEADERS = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}
```

### Perform a Semantic Search
Using the `/v1/search` endpoint, we can find images purely based on their visual meaning or text content.

```python
query = "a dog running in a park"
params = {
    "keyword": query,
    "limit": 3
}

response = requests.post(f"{API_URL}/search", json=params, headers=HEADERS)

if response.status_code == 200:
    results = response.json().get('data', [])
    print(f"Found {len(results)} results for: '{query}'\n")
    
    # Note: Because the API returns delivery URLs, we can actually display them in the notebook!
    html_str = "<div style='display:flex; gap: 10px;'>"
    for asset in results:
        # The API likely returns a render URL or we can construct one
        asset_url = asset.get('url')
        if asset_url:
             html_str += f"<img src='{asset_url}' style='width: 200px; border-radius: 8px;' />"
    html_str += "</div>"
    
    if results:
        display(HTML(html_str))
else:
    print(f"Error: {response.status_code} - {response.text}")
```
