# Picsha AI: API Basics
This notebook demonstrates how to connect to the Picsha AI backend and fetch your intelligent media assets.

```python
import os
import requests
import json

# Configuration
API_URL = "https://api.picsha.ai/v1"
API_KEY = os.getenv("PICSHA_API_KEY", "YOUR_API_KEY_HERE")

HEADERS = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}
print("Ready to make requests!")
```

### 1. Retrieve Assets
Let's fetch a list of uploaded images to see what data the API returns.

```python
response = requests.get(
    f"{API_URL}/assets", 
    headers=HEADERS, 
    params={"type": "image", "limit": 3}
)

if response.status_code == 200:
    data = response.json()
    print(f"Found {len(data.get('data', []))} assets.\n")
    if data.get('data'):
        # Show the first asset's structure
        print(json.dumps(data['data'][0], indent=2))
else:
    print(f"Error: {response.status_code} - {response.text}")
```
